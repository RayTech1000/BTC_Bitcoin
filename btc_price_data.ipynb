```python
today = date.today()
d1 = today.strftime("%Y-%m-%d")
end_date = d1
d2 = date.today() - timedelta(days=365)
d2 = d2.strftime("%Y-%m-%d")
start_date = d2

#Apple Stock Data
apple_stocks_data = yf.download('AAPL',
                      start=start_date, 
                      end=end_date, 
                      progress=False)
print(apple_stocks_data.head())
```

```python
today = date.today()
d1 = today.strftime("%Y-%m-%d")
end_date = d1
d2 = date.today() - timedelta(days=365)
d2 = d2.strftime("%Y-%m-%d")
start_date = d2

#Bitcoin data
btc_data = yf.download('BTC-USD', 
                      start=start_date, 
                      end=end_date, 
                      progress=False)
```

```python
btc_data["Date"] = btc_data.index
btc_data = btc_data[["Date", "Open", "High", "Low", "Close", "Adj Close", "Volume"]]
btc_data.reset_index(drop=True, inplace=True)
```

```python
fig1 = px.bar(data_frame = btc_data, x = "Date", y = "Volume", title = '$Bitcoin Volume 2022-2023', width=800, height=400)
fig1.show()
```

```python
fig2 = px.scatter(data_frame = btc_data, x = 'Date', y="High", size="High", color= "High", title= "$Bitcoin High Price 2022-2023", trendline = 'ols', width=1000, height=500)
fig2.show()
```

```python
fig2_1 = px.scatter(data_frame = btc_data, x = 'Date', y="Low", size="Low", color= "Low", title = "$Bitcoin Low Price 2022-2023", trendline = 'ols', width=1000, height=500)
fig2_1.show()
```

```python
import plotly.graph_objects as go
fig3 = go.Figure(data=[go.Candlestick(x=btc_data["Date"],
                                        open=btc_data["Open"], 
                                        high=btc_data["High"],
                                        low=btc_data["Low"], 
                                        close=btc_data["Close"])])
fig3.update_layout(title = "$Bitcoin Price 2022-2023",
                     width=1000, height=500,
                     xaxis_rangeslider_visible=True)
fig3.show()
```
